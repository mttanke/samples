## Configure model + region

We use Claude Sonnet 4.5 on Amazon Bedrock in `us-east-1`. Make sure you have [model access](https://docs.aws.amazon.com/bedrock/latest/userguide/model-access-modify.html) enabled.

```python
import os

MODEL_ID = "global.anthropic.claude-opus-4-8"  # Claude Sonnet 4.5
REGION   = "us-east-1"
os.environ["STRANDS_MODEL_ID"] = MODEL_ID
os.environ["AWS_REGION"] = REGION
print(f"Model:  {MODEL_ID}")
print(f"Region: {REGION}")
```

## Install Strands (first time only)

```python
%pip install -q strands-agents strands-agents-tools
```

## The agent

```python
from strands import Agentfrom strands_tools import shellagent = Agent(    model=MODEL_ID,    tools=[shell],    system_prompt="You are a helpful assistant. Use the shell tool when useful. Be concise.",)
```

## Ask it things

```python
agent("hello, introduce yourself")
```

```python
agent("what files are in this directory?")
```

### What just happened?Claude decided to call `shell(ls)` on its own. That's Strands' model-first philosophy.
