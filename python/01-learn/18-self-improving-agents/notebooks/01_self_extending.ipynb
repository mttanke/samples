```python
import os

MODEL_ID = "global.anthropic.claude-opus-4-8"  # Claude Sonnet 4.5
os.environ["AWS_REGION"] = "us-east-1"
os.makedirs("tools", exist_ok=True)
```

## The magic: `load_tools_from_directory=True`

```python
from strands import Agentfrom strands_tools import shell, file_writeagent = Agent(    model=MODEL_ID,    tools=[shell, file_write],    load_tools_from_directory=True,   # 👈 THE MAGIC    system_prompt=(        "You can create new tools by writing Python files to ./tools/. "        "Decorate them with @tool from the strands package. "        "They become available instantly."    ),)
```

## Ask the agent to build a tool

```python
agent("create a tool called password_generator that makes random passwords, then use it to make a 20-char password")
```
